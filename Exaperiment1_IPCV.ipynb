```python
img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

img_gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
```

```python
plt.figure(figsize=(12,5))

plt.subplot(1,2,1)
plt.imshow(img_rgb)
plt.title("Original Image")
plt.axis("off")

plt.subplot(1,2,2)
plt.imshow(img_gray, cmap="gray")
plt.title("Gray Image")

plt.tight_layout()
plt.show()
```

```python
#  Convert grayscale to BGR (3-channel) so dimensions match
img_gray_3channel = cv2.cvtColor(img_gray, cv2.COLOR_GRAY2BGR)

#  Subtract the images
subtract = cv2.subtract(img_rgb, img_gray_3channel)

#  convert the result from BGR to RGB (e.g.Matplotlib)
subtract_rgb = cv2.cvtColor(subtract, cv2.COLOR_BGR2RGB)
```

```python
plt.imshow(subtract_rgb)
plt.title("Subtract Image")
plt.axis("off")
plt.show()
```
